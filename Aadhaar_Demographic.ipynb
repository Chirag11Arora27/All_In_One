```python
state_demo = (
    df.groupby('state')['total_demographic_updates']
        .sum()
        .sort_values(ascending=False)
        .head(10)
)

plt.figure(figsize=(10,5))
state_demo.plot(kind='bar')
plt.title("Top 10 States by Aadhaar Demographic Updates")
plt.xlabel("State")
plt.ylabel("Total Updates")
plt.show()
```
